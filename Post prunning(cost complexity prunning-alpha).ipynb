```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
data
```

```python
cancer=pd.DataFrame(np.c_[data["data"],data["target"]], columns=np.append(data["feature_names"],["Target"]))
cancer
```

### 1st method get X and y

```python
X=cancer.drop(["Target"], axis=1)
y=cancer['Target']
```

```python
y
```

### 2nd method to get X and y

```python
X1,Y1=load_breast_cancer(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(random_state=0)
model=clf.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```

```python
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(clf,filled=True)
```

```python
prun=clf.cost_complexity_pruning_path(X_train,y_train)
ccp_alpha, impurity= prun.ccp_alphas, prun.impurities
```

```python
ccp_alpha
```

```python
ccp=[]
for i in ccp_alpha:
    clf=DecisionTreeClassifier(random_state=0, ccp_alpha=i)
    clf.fit(X_train,y_train)
    ccp.append(clf)
```

```python
ccp
```

```python
train_score=[i.score(X_train,y_train) for i in ccp]
test_score=[i.score(X_test, y_test) for i in ccp]
```

```python
fig, ax=plt.subplots()
ax.set_xlabel("Alphas")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy versus ccp-Alphas graph ")
ax.plot(ccp_alpha,train_score, marker='o',label="train",drawstyle="steps-post")
ax.plot(ccp_alpha,test_score, marker='o', label="test",drawstyle="steps-post")
ax.legend()
plt.show()
```
